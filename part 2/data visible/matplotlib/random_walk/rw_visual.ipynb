```python
#绘制随机漫步的所有点

import matplotlib.pyplot as plt

from random_walk import RandomWalk

#只要程序处于活动状态，就不断的模拟随机漫步
while True:
    rw=RandomWalk(50000)  #可以增加点的数量
    rw.fill_walk()

    #设置绘图窗口的尺寸
    plt.figure(figsize=(10,6)) #函数figure()用于指定图表的宽度、高度、分辨率和背景色
                               #给形参figsize提供一个元组，指出绘图窗口的尺寸，单位为英寸

    point_numbers=list(range(rw.num_points))

    #隐藏坐标轴
    current_axes = plt.axes()
    current_axes.get_xaxis().set_visible(False)
    current_axes.get_yaxis().set_visible(False)
    #注意事项：1、必须先赋值给一个变量再调用
    #         2、必须在scatter()前使用

    plt.scatter(rw.x_values,rw.y_values,c=point_numbers,cmap=plt.cm.Blues,
                edgecolors='None',s=1)   #最开始的点设置为浅色，后面的点设置为深色
                                          #由于各点按顺序绘制，所以给参数c指定的列表只需包含数字1~5000
    
    #突出起点和终点
    plt.scatter(0,0,c='green',edgecolors='none',s=30)                           #用绿色绘制点(0,0)——即起点
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',edgecolors='none',s=30) #用红色绘制终点

    plt.show()

    

    keep_running=input("Make another walk? (y/n): ")
    if keep_running=='n':
        break
```
