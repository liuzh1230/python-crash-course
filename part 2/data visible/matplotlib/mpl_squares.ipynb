```python
import matplotlib.pyplot as plt #模块pyplot包含很多生成图表的函数

#可访问matplotlib.org的示例画廊

squares=[1,4,9,16,25]
plt.plot(squares,linewidth=5)  #参数linewidth决定了plot()绘制的线条粗细

#设置图表标题，并给坐标轴加上标签
plt.title("Square numbers",fontsize=24)  #函数title()为图表指定标题，fontsize指定了字体大小
plt.xlabel("Value",fontsize=20)
plt.ylabel("Square of value",fontsize=14)

#设置刻度标记的大小
plt.tick_params(axis="both",labelsize=14)  #用于自定义刻度标记，其中axis指定要设置的坐标轴
                                           #这里labelsize用于设置标记大小
                                           #其他标记方法可以自行搜索

plt.show() #打开matplotlib查看器，显示绘制的图形
```

```python
#在上一张图中，我们注意到x-y对应关系出现问题(如4-25)
#这是因为在向plot()提供一系列数字时，其假设第一个数据点对应的x坐标为0
#为改变默认值，我们可以向其提供输入值(x值)和输出值(y值)
import matplotlib.pyplot as plt  #导入模块pyplot

input_values=[1,2,3,4,5]
squares=[1,4,9,16,25]
plt.plot(input_values,squares,linewidth=3)  

#设置图表标题，并给坐标轴加上标签
plt.title("Square numbers",fontsize=24)  
plt.xlabel("Value",fontsize=20)
plt.ylabel("Square of value",fontsize=14)

#设置刻度标记的大小
plt.tick_params(axis="both",labelsize=14)  

plt.show() 
```
